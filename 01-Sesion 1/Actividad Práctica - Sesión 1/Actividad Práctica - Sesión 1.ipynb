Las etiquetas, en la columna `label` se expresan en valores textuales. Hay casos en que las etiquetas deben expresarse en valores enteros. A continuación se creará una nueva columna `label_num` que expresa las etiquetas como números: "weak" = 0; "medium" = 1; "strong" = 2. 

```python
# Define un diccionario que mapea las etiquetas textuales a valores numéricos
label_mapping = {"weak": 0, "medium": 1, "strong": 2}

# Crea la nueva columna 'label_num' aplicando el mapeo
rockyou_df['label_num'] = rockyou_df['label'].map(label_mapping)

# Muestra el DataFrame actualizado
print(tabulate(rockyou_df.head(20), headers=rockyou_df.columns, 
               showindex=False,  tablefmt="pretty"))
```

Como se puede ver, el dataset `rockyou` es masivo: más de 14 millones de contraseñas. Procesar toda esa información puede ser muy tardado. Para efectos de este taller, se traajará solamente con una pequeña porción de los datos de apenas el 1%. Aunque se estará trabajando con un pequeño subconjunto de los datos, los análisis presentados se pueden realizar a la totalidad de `rockyou` en caso de contar con el tiempo y el poder de cómputo adecuado. Se debe destacar que los resultados de los análsis realizados para el subconjunto del 1% pueden no ser los mismos que para el 100% de los datos.

Seguidamente se seleccionarán las contraseñas del subconjunto reducido manteniendo las mismas proporciones que tienen los datos originales.

```python
# Quita la columna "label_num" del DataFrame rockyou_df y asigna el resultado a X.
X = rockyou_df.drop("label_num", axis=1)

# Asigna la columna "label_num" del DataFrame rockyou_df a y.
y = rockyou_df["label_num"]

# Divide los conjuntos de datos X y y en conjuntos de datos de entrenamiento y prueba,
# donde X_val y y_val serán utilizados para la validación con un tamaño de prueba del 1%
# y una semilla aleatoria de 42 para la reproducibilidad.
_, X_val, _, y_val = train_test_split(X, y, test_size=0.01, random_state=42)

# Combina X_val y y_val en un solo DataFrame manteniendo la estructura de filas y columnas de X_val
rockyou_df = pd.concat([X_val, pd.DataFrame({'y_val': y_val})], axis=1)

# Ajustar el nombre de la columna "y_val" por "label_num"
rockyou_df = rockyou_df.rename(columns={'y_val': 'label_num'})

# Resetar los indices para evitar conflictos luego
rockyou_df.reset_index(drop=True, inplace=True)

# Obtiene el número de filas y columnas del DataFrame generado
num_filas, num_columnas = rockyou_df.shape

# Imprime un mensaje amigable para el usuario mostrando las dimensiones del nuevo dataset
print(f"El dataset reducido tiene {num_filas} filas y {num_columnas} columnas.")
```

2. ¿Cuál es el conjunto de caracteres usados en cada uno de los grupos de contraseñas del inciso anterior?

Para determinar el conjunto de caracteres utilizados en cada uno de los grupos de contraseñas (según su longitud) del inciso anterior, podemos seguir estos pasos:

1. Filtrar el DataFrame para obtener solo las contraseñas con la longitud mínima, máxima y media.
2. Iterar sobre las contraseñas en cada grupo y extraer los caracteres únicos.
3. Mostrar el conjunto de caracteres únicos para cada grupo.

```python
# Obtener las contraseñas con longitud mínima, máxima y media
min_length_passwords = rockyou_df[rockyou_df['password_length'] == password_length_stats['min']]['password']
max_length_passwords = rockyou_df[rockyou_df['password_length'] == password_length_stats['max']]['password']
mean_length_passwords = rockyou_df[rockyou_df['password_length'] == round(password_length_stats['mean'])]['password']

# Función para encontrar el conjunto de caracteres únicos en una serie de contraseñas
def unique_characters(passwords):
    unique_chars = set()
    for password in passwords:
        unique_chars.update(password)
    return unique_chars

# Mostrar el conjunto de caracteres únicos para cada grupo de contraseñas
print(f"Conjunto de caracteres en contraseñas de longitud mínima. Tamaño: {len(unique_characters(min_length_passwords))} \n {unique_characters(min_length_passwords)}")
print(f"Conjunto de caracteres en contraseñas de longitud máxima. Tamaño: {len(unique_characters(max_length_passwords))} \n {unique_characters(max_length_passwords)}")
print(f"Conjunto de caracteres en contraseñas de longitud media. Tamaño: {len(unique_characters(mean_length_passwords))} \n {unique_characters(mean_length_passwords)}")
```

### Aplicar un algoritmo de agrupamiento para agrupar las contraseñas.

Los algoritmos de agrupamiento permiten determinar aquellos datos con características semejantes puesto que caen en un mismo grupo. También ayudan a determinar si los datos son "separables". Para que estos algoritmos funcionen, se debe indicar el número de grupos a crear. No existe un método para saber a priori cuál es el mejor número de grupos y a fin de cuentas, el número óptimo de grupos lo determinará la tarea que se está realizando. Una manera para poder determinar un buen número de grupos es el llamado ["Método del Codo"](https://www.geeksforgeeks.org/elbow-method-for-optimal-value-of-k-in-kmeans/) (Elbow).

Primero, se incializan las estructuras de datos a emplear y se separan los datos. En este caso, se realizará el agrupamiento considerando solamente los valores numéricos de los datos. Luego, estos valores numéricos deben normalizarse para queden todos dentro del mismo rango.

```python
print(tabulate(rockyou_df.head(20), headers='keys', showindex=False, 
               tablefmt='beauty'))
```

En este punto ya se puede comenzar a procesar para obtener el número óptimo de grupos a partir de los datos normalizados. Los grupos obtenidos se evaluarán de acuerdo a 3 métricas:

- **Coeficiente de inercia:** El coeficiente `inertia` (inercia) es una medida utilizada en algoritmos de agrupamiento, específicamente en técnicas como K-Means. También es conocida como "suma de cuadrados dentro del grupo" (SSW). La inercia mide cómo están distribuidos los puntos dentro de cada clúster. En otras palabras, representa la suma de las distancias al cuadrado de cada punto al centroide más cercano de su clúster. El objetivo principal de los algoritmos de agrupamiento, como K-Means, es minimizar esta inercia. Cuando la inercia es baja, significa que los puntos dentro de cada clúster están cercanos entre sí y, por lo tanto, el agrupamiento es más coherente. La inercia se utiliza comúnmente como criterio para evaluar la calidad de los agrupamientos obtenidos con diferentes números de clústeres. Se espera que a medida que el número de clústeres aumenta, la inercia disminuya, pero hay un punto donde agregar más clústeres no mejora significativamente la coherencia de los agrupamientos, lo que se conoce como el "elbow" (codo) en el gráfico de la inercia en función del número de clústeres. En este punto, se suele elegir el número óptimo de clústeres para el problema en cuestión.

- **Coeficiente de Silhouettete:** El coeficiente de Silhouette es una medida de la calidad y coherencia de los agrupamientos obtenidos a través de algoritmos de clustering, como K-Means, jerárquico, DBSCAN, entre otros. Esta métrica proporciona una forma de evaluar la separación entre los clústeres así como la cohesión de los puntos dentro de cada clúster.

  El coeficiente de Silhouette se calcula para cada punto individual y proporciona un valor entre -1 y 1. 

  - Un valor cercano a 1 indica que el punto está bien clasificado en su clúster y está lejos de los puntos en los otros clústeres.

  - Un valor cercano a 0 indica que el punto está cerca del límite de decisión entre dos clústeres.

  - Un valor cercano a -1 indica que el punto puede haber sido asignado al clúster incorrecto.

  El coeficiente de Silhouette global se calcula como el promedio de los coeficientes de Silhouette de todos los puntos en el conjunto de datos. Un coeficiente de Silhouette global alto sugiere que los clústeres están bien definidos y separados entre sí.

  El coeficiente de Silhouette es útil para determinar el número óptimo de clústeres en un conjunto de datos, ya que puede ayudar a identificar el punto en el que la estructura del clúster es más fuerte. Sin embargo, se debe tener en cuenta que esta métrica no debe utilizarse como la única medida para evaluar la calidad de los clústeres, y es recomendable complementarla con otras métricas y técnicas de validación de clústeres.

  **En este ejemplo no se calcula porque puede resultar muy tardado, incluso para conjuntos con pocos datos.**

- **Índice Davies-Bouldin:** El índice de Davies-Bouldin es una métrica utilizada para evaluar la calidad de los clústeres obtenidos a través de algoritmos de agrupamiento. Fue propuesto por David L. Davies y D. W. Bouldin en 1979.

  El índice de Davies-Bouldin se calcula considerando la dispersión dentro de cada clúster y la dispersión entre los clústeres. Un valor más bajo del índice de Davies-Bouldin indica una mejor partición de los datos, donde los clústeres son más densos y están más separados unos de otros.

  Para calcular el índice de Davies-Bouldin, se realiza lo siguiente para cada par de clústeres:

  1. Se calcula la dispersión dentro del clúster como la distancia promedio entre cada punto del clúster y el centroide del mismo.
  2. Se calcula la dispersión entre los clústeres como la distancia entre los centroides de los clústeres.
  3. Se calcula el índice de Davies-Bouldin para el par de clústeres como la suma de las dispersiones dentro del clúster y entre los clústeres, dividida por la dispersión dentro del clúster más grande.

  El índice de Davies-Bouldin se calcula como el promedio de los índices para todos los pares de clústeres. Un valor más bajo indica una mejor separación entre los clústeres y, por lo tanto, una mejor calidad de los clústeres.

  Esta métrica es útil para comparar diferentes algoritmos de agrupamiento y diferentes configuraciones de parámetros para un mismo algoritmo. Sin embargo, al igual que otras métricas de evaluación de clústeres, no debe considerarse de forma aislada y es recomendable utilizarla junto con otras métricas y técnicas de validación de clústeres para obtener una evaluación más completa.

```python
# Lista para almacenar los valores de la inercia, 
# coeficiente de Silhouette y Davies-Bouldin score
inertia_values = []
silhouette_scores = []
davies_bouldin_scores = []

# Calcular los agrupamientos y la inercia y Davies-Bouldin score para cada valor de k
for k in k_values:
    print(f"Analizando agrupamiento para k={k}....")
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia_values.append(kmeans.inertia_)
    
    cluster_labels = kmeans.labels_
    davies_bouldin_avg = davies_bouldin_score(X_scaled, cluster_labels)
    davies_bouldin_scores.append(davies_bouldin_avg)
```

A partir de los agrupamientos obtenidos, se debe buscar el mejor numero k

```python
# Imprimir el Davies-Bouldin score para cada valor de k
for k, score in zip(k_values, davies_bouldin_scores):
    print(f'Davies-Bouldin score para k={k}: {score}')

# Determinar el número óptimo de clusters basado en las métricas
best_k_inertia = np.argmin(inertia_values) + 2  # Índice mínimo + 2 porque k_values comienza en 2
best_k_davies_bouldin = np.argmin(davies_bouldin_scores) + 2

print(f'Número óptimo de clusters según la inercia: {best_k_inertia}')
print(f'Número óptimo de clusters según el puntaje de Davies-Bouldin: {best_k_davies_bouldin}')
```

Una vez que se ha obtenido de manera analítica el mejor número de grupos, se muestran de manera visual para tener un mejor entendimiento.

Una vez creados los conjuntos, se evaluarán varios clasificadores para determinar el que mejor desempeño obtiene en el conjunto de entrenamiento y de validación. El que mejor desempeño obtenga, será empleado para asignar una fortaleza a las contraseñas del set de prueba. Para ello se usará un "pipeline" de python. En Python, un "pipeline" es una secuencia de pasos que se ejecutan de manera ordenada, donde la salida de un paso se convierte en la entrada del siguiente. En el contexto de aprendizaje automático, un pipeline es una forma conveniente de encadenar múltiples transformaciones de datos y modelos en un flujo de trabajo único y coherente.

Los pipelines son especialmente útiles porque encapsulan todo el flujo de trabajo en un objeto único que se puede ajustar, entrenar y evaluar de manera más conveniente. También ayudan a prevenir errores comunes, como la filtración de datos del conjunto de prueba en el preprocesamiento. Además, los pipelines facilitan la reproducibilidad del flujo de trabajo al asegurar que los mismos pasos se apliquen de manera consistente a todos los conjuntos de datos.

El pipeline usado en esta sección del código emplea diversos algoritmos de clasificación. Estos algoritmos fueron seleccionados por ser los más conocidos de cada familia. Los algoritmos seleccionados fueron:

1. **K-Nearest Neighbors (KNN)**:
   - Funcionamiento: KNN funciona asignando una etiqueta a un punto de datos basado en la mayoría de las etiquetas de sus vecinos más cercanos. Utiliza una métrica de distancia, como la distancia euclidiana, para calcular la similitud entre los puntos de datos.
   - Ventajas: Es simple y fácil de entender. No requiere entrenamiento previo y puede ser eficiente para conjuntos de datos pequeños.
   - Desventajas: Puede ser computacionalmente costoso para grandes conjuntos de datos. Sensible a la elección de la métrica de distancia y al número de vecinos.

2. **Random Forest**:
   - Funcionamiento: Random Forest crea múltiples árboles de decisión durante el entrenamiento y combina sus predicciones para obtener una predicción final. Cada árbol se entrena en una muestra aleatoria del conjunto de datos y en una selección aleatoria de características.
   - Ventajas: Tiende a tener un buen rendimiento en una variedad de conjuntos de datos. Es menos propenso al sobreajuste en comparación con un árbol de decisión único.
   - Desventajas: Puede ser más lento y requerir más recursos computacionales que otros algoritmos más simples como KNN.

3. **Support Vector Machine (SVM)**:
   - Funcionamiento: SVM busca encontrar el hiperplano óptimo que mejor separa las clases en el espacio de características. Puede manejar conjuntos de datos lineales y no lineales utilizando diferentes funciones de kernel.
   - Ventajas: Puede ser efectivo en espacios de alta dimensionalidad y es versátil debido a la capacidad de usar diferentes funciones de kernel.
   - Desventajas: Puede ser sensible a la elección del parámetro de regularización y puede ser computacionalmente costoso para conjuntos de datos grandes.

4. **Logistic Regression**:
   - Funcionamiento: Logistic Regression modela la probabilidad de que una instancia pertenezca a una clase particular utilizando la función logística. Estima los coeficientes de regresión que maximizan la verosimilitud de los datos observados.
   - Ventajas: Es simple y rápido de entrenar. Proporciona una probabilidad de pertenencia a una clase.
   - Desventajas: Puede no funcionar bien si las clases no están bien separadas linealmente. No es adecuado para conjuntos de datos no lineales.

5. **Neural Network (MLPClassifier)**:
   - Funcionamiento: MLPClassifier utiliza una arquitectura de red neuronal multicapa para aprender una función que mapea las características de entrada a las etiquetas de salida. Utiliza el algoritmo de retropropagación para ajustar los pesos de la red durante el entrenamiento.
   - Ventajas: Puede capturar relaciones complejas entre las características y las etiquetas. Es flexible y puede adaptarse a una variedad de problemas de aprendizaje automático.
   - Desventajas: Puede ser propenso al sobreajuste, especialmente en conjuntos de datos pequeños. Requiere una cantidad significativa de datos y recursos computacionales para entrenar modelos profundos.

Para evaluar el desempeño de los algoritmos propuestos se emplearon diversas métricas:

1. **Accuracy (Precisión)**:
   - Descripción: Mide la fracción de predicciones correctas realizadas por el modelo sobre el total de predicciones.
   - Fórmula: $$ \text{Accuracy} = \frac{\text{Verdaderos positivos} + \text{Verdaderos negativos}}{\text{Total de muestras}} $$
   - Interpretación: Cuanto más alto sea el valor de precisión, mejor es el rendimiento del modelo en términos de predicciones correctas en general.

2. **Precision (Precisión)**:
   - Descripción: Mide la fracción de instancias positivas identificadas correctamente por el modelo entre todas las instancias que el modelo clasificó como positivas.
   - Fórmula: $$ \text{Precision} = \frac{\text{Verdaderos positivos}}{\text{Verdaderos positivos} + \text{Falsos positivos}} $$
   - Interpretación: Una alta precisión indica que el modelo es bueno para predecir instancias positivas, minimizando los falsos positivos.

3. **Recall (Exhaustividad o Sensibilidad)**:
   - Descripción: Mide la fracción de instancias positivas que fueron correctamente identificadas por el modelo entre todas las instancias que son realmente positivas.
   - Fórmula: $$ \text{Recall} = \frac{\text{Verdaderos positivos}}{\text{Verdaderos positivos} + \text{Falsos negativos}} $$
   - Interpretación: Cuanto mayor sea el valor de recall, mejor será el modelo para capturar todas las instancias positivas, minimizando los falsos negativos.

4. **F1 Score**:
   - Descripción: Es una medida que combina precisión y recall en un solo valor, calculado como la media armónica de ambos.
   - Fórmula: $$F1 \text{ Score} = 2 \times \frac{\text{Precisión} \times \text{Exhaustividad}}{\text{Precisión} + \text{Exhaustividad}} $$
   - Interpretación: El F1 Score es útil cuando se desea encontrar un equilibrio entre precisión y recall. Es especialmente útil cuando las clases están desequilibradas.
