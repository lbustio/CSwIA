Para hacer más modular el código, se definirá una función que lee un archivo `txt` y devolverá un dataframe.

```python
def leer_datos(path):
    # Cargar el archivo en una lista de strings
    with open(path, 'r', encoding='unicode_escape') as file:
        lines = file.readlines()
    
    # Convertir la lista en un DataFrame de Pandas
    result_df = pd.DataFrame({'txt': lines})

    return result_df
```

#### 2. Frecuencia de palabras clave

Para este punto, que implica analizar la frecuencia de palabras clave en los mensajes de phishing y legítimos, se pueden seguir estos pasos:

1. Identificar palabras clave que puedan estar asociadas con phishing.
2. Contar la frecuencia de estas palabras clave en los mensajes de phishing y legítimos.
3. Comparar las frecuencias de las palabras clave entre los dos tipos de correos electrónicos.

```python
# Definir palabras clave asociadas con phishing.
# Estas palabras clave se pueden determinar de manera automática
# mediante otros algoritmos.
palabras_clave_phishing = ['urgente', 'contraseña', 'verificación', 'oferta', 'confirmación']

# Calcular la frecuencia de las palabras clave en los mensajes de phishing
frecuencia_phishing = {}
for palabra in palabras_clave_phishing:
    frecuencia_phishing[palabra] = data_df[data_df['label'] == 'phish']['txt'].str.lower().str.count(palabra).sum()

# Calcular la frecuencia de las palabras clave en los mensajes legítimos
frecuencia_legit = {}
for palabra in palabras_clave_phishing:
    frecuencia_legit[palabra] = data_df[data_df['label'] == 'legit']['txt'].str.lower().str.count(palabra).sum()

# Convertir los diccionarios en listas de tuplas para tabulate
tabla_phishing = list(frecuencia_phishing.items())
tabla_legit = list(frecuencia_legit.items())

# Mostrar los resultados en forma de tabla
print("Frecuencia de palabras clave para phishing:")
print(tabulate(tabla_phishing, headers=['Palabra clave', 'Frecuencia']))
print("\nFrecuencia de palabras clave para legítimos:")
print(tabulate(tabla_legit, headers=['Palabra clave', 'Frecuencia']))
```

A partir de los resultados de las frecuencias de palabras clave para phishing y correos legítimos, se pueden hacer algunas observaciones:

1. **Palabras clave comunes:** Las palabras clave como "urgente", "contraseña", "verificación", "oferta" y "confirmación" aparecen en ambos tipos de correos electrónicos, aunque con diferentes frecuencias.

2. **Diferencias en frecuencias:** Las frecuencias de estas palabras clave varían entre los correos de phishing y legítimos. Por ejemplo, la palabra clave "urgente" aparece en 10 correos de phishing pero no aparece en ningún correo legítimo, lo que sugiere que "urgente" podría ser una señal de phishing. Del mismo modo, otras palabras clave como "contraseña", "oferta" y "confirmación" aparecen con mayor frecuencia en los correos de phishing en comparación con los correos legítimos.

3. **Verificación:** Curiosamente, la palabra clave "verificación" aparece con una frecuencia de 7 en los correos de phishing y con una frecuencia de 3 en los correos legítimos. Esto podría indicar que la palabra "verificación" es más común en los correos de phishing, pero también está presente en algunos correos legítimos.

Estos resultados sugieren que ciertas palabras clave pueden ser indicadores útiles para distinguir entre correos de phishing y correos legítimos. ### Análisis de Agrupamiento

Una vez entendidos los patrones preliminares de los datos, se deben realizar análisis más complejos y profundos para entender mejor la naturaleza de estos datos. Un ejemplo de ello es el Análisis de Agrupamiento, que ya se vió en la sesión anterior.

Para el agrupamiento de datos se usará el algoritmo `KMeans` ya estudiado, por su facilidad de uso y practicidad. KMeans no funciona con datos textuales, por lo que se deben codificar los textos a valores numéricos, en los que los vectores numéricos mantengan la información de los datos y que por lo tanto, se puedan utilizar en los análisis.

La forma que se empleará en esta sesión para convertir los datos textuales en vectores numéricos equivalentes de características es TF-IDF.

TF-IDF (Term Frequency-Inverse Document Frequency) es una medida numérica que se utiliza comúnmente en el procesamiento de lenguaje natural y la recuperación de información para evaluar la importancia de una palabra en un documento en relación con una colección de documentos más amplia. 

La idea básica detrás de TF-IDF es que una palabra es relevante para un documento si aparece con frecuencia en ese documento (frecuencia de término - TF) pero no es muy común en otros documentos de la colección (frecuencia inversa de documento - IDF).

La fórmula para calcular TF-IDF es:

<div align="center">

  $\text{TF-IDF}(t, d, D) = \text{tf}(t, d) \times \text{idf}(t, D)$

  </div>

Donde:
- $\text{TF}(t, d)$ es la frecuencia de término de un término $t$ en un documento $d$.
- $\text{IDF}(t, D)$ es la frecuencia inversa de documento de un término $t$ en una colección de documentos $D$.

La idea es que cuanto mayor sea el valor de TF-IDF para un término en un documento específico, más importante es ese término para ese documento en particular. Esto se debe a que el término aparece con frecuencia en el documento (alta TF) pero no es común en otros documentos de la colección (alta IDF).


```python
# Preprocesamiento de datos
tfidf_vectorizer = TfidfVectorizer(stop_words=list(stop_words_spanish))
tfidf_matrix = tfidf_vectorizer.fit_transform(data_df['txt'])  # correos es tu lista de correos electrónicos
```

Con los datos vectorizados se puede proceder al agrupamiento. En este punto, y tratándose de un problema de aprendizaje no supervisado, se sospecha que los datos se agrupen en 2 grupos, pero esto debe verificarse. Para ello, se debe obtener el número óptimo de grupos, lo cual se logra usando el método del codo.

```python
# Lista para almacenar los valores de la inercia, coeficiente de Silhouette y Davies-Bouldin score
inertia_values = []
silhouette_scores = []
davies_bouldin_scores = []

k_values = range(2,11,1)

# Calcular los agrupamientos y la inercia y Davies-Bouldin score para cada valor de k
for k in k_values:
    print(f"Analizando agrupamiento para k={k}....")
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(tfidf_matrix)
    inertia_values.append(kmeans.inertia_)
    
    cluster_labels = kmeans.labels_
    silhouette_avg = silhouette_score(tfidf_matrix, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    
    davies_bouldin_avg = davies_bouldin_score(tfidf_matrix.toarray(), cluster_labels)
    davies_bouldin_scores.append(davies_bouldin_avg)
```

A partir de los coeficientes anteriores, se debe obtener el número óptimo de clusters para este probelma.

```python
# Imprimir el Davies-Bouldin score y el coeficiente de Silhouette para cada valor de k
for k, davies_bouldin_score, silhouette_score in zip(k_values, davies_bouldin_scores, silhouette_scores):
    print(f'Para k={k}: Davies-Bouldin score: {davies_bouldin_score}, Coeficiente de Silhouette: {silhouette_score}')

# Determinar el número óptimo de clusters basado en las métricas
best_k_inertia = np.argmin(inertia_values) + 2  # Índice mínimo + 2 porque k_values comienza en 2
best_k_davies_bouldin = np.argmin(davies_bouldin_scores) + 2
best_k_silhouette = np.argmax(silhouette_scores) + 2

print(f'Número óptimo de clusters según la inercia: {best_k_inertia}')
print(f'Número óptimo de clusters según el puntaje de Davies-Bouldin: {best_k_davies_bouldin}')
print(f'Número óptimo de clusters según el coeficiente de Silhouette: {best_k_silhouette}')
```

Una vez que se ha obtenido de manera analítica el mejor número de grupos, se muestran de manera visual para tener un mejor entendimiento.

```python
# Inicializar figuras
fig, axes = plt.subplots(1, 1, figsize=(12, 6))

# Graficar la inercia en función del número de clusters
axes.plot(k_values, inertia_values, marker='o', label='Inercia', color='black')
axes.axvline(x=k_values[np.argmin(inertia_values)], 
             color='red', linestyle='--', label='Número Óptimo de Grupos (Inercia)')
axes.axvline(x=k_values[np.argmin(davies_bouldin_scores)], 
             color='blue', linestyle='--', label='Número Óptimo de Grupos (Davies-Bouldin)')
axes.axvline(x=k_values[np.argmax(silhouette_scores)], 
             color='green', linestyle='--', label='Número Óptimo de Grupos (Silhouette)')
axes.set_xlabel('Número de Clusters (k)')
axes.set_ylabel('Inercia')
axes.set_title('Método del Codo para Encontrar el Número Óptimo de Clusters')

# Ajustar el diseño de los gráficos
plt.tight_layout()
axes.legend(loc='upper right')

# Mostrar el gráfico
plt.show()
```

Una vez obtenido el número óptimo de grupos, se realiza el agrupamiendo de los datos usando el número óptimo de grupos sugerido.

Tambien se pueden usar 2 grupos para ver la diferencia entre un agrupamiento estimado y un agrupamiendo duro.
